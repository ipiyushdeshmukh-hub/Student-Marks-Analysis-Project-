```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/StudentsPerformance.csv")
df.head()
```

```python
plt.figure(figsize=(10,5))
plt.hist(df['math score'])
plt.title("Math Score Distribution")
plt.xlabel("Score")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(df['reading score'], df['writing score'])
plt.title("Reading vs Writing Score")
plt.xlabel("Reading Score")
plt.ylabel("Writing Score")
plt.show()
```

```python
df[['math score', 'reading score', 'writing score']].mean()
```

```python
df[['math score', 'reading score', 'writing score']].max()
```

```python
df[['math score', 'reading score', 'writing score']].min()
```

```python
top_students = df.head(10)

plt.figure(figsize=(12,6))
plt.bar(top_students.index, top_students['math score'], color='skyblue')
plt.title("Top 10 Students - Math Score")
plt.xlabel("Student Index")
plt.ylabel("Math Score")
plt.xticks(top_students.index)
plt.show()
```

```python
subjects = ['math score', 'reading score', 'writing score']
averages = df[subjects].mean()

plt.figure(figsize=(8,5))
plt.bar(subjects, averages, color=['red','green','blue'])
plt.title("Average Marks Per Subject")
plt.xlabel("Subjects")
plt.ylabel("Average Score")
plt.show()
```

```python
# Gender-wise Analysis
df.groupby('gender')[['math score','reading score','writing score']].mean()
```

```python
gender_avg = df.groupby('gender')[['math score','reading score','writing score']].mean()

gender_avg.plot(kind='bar', figsize=(8,5))
plt.title("Gender-wise Average Marks")
plt.ylabel("Average Score")
plt.xticks(rotation=0)
plt.show()
```

```python
df.groupby('parental level of education')[['math score','reading score','writing score']].mean()
```

```python
df.groupby('parental level of education')['math score'].mean().plot(kind='bar', figsize=(10,5))
plt.title("Parental Education vs Math Score")
plt.ylabel("Average Math Score")
plt.show()
```

```python
df.groupby('lunch')['math score'].mean()
```

```python
df.groupby('lunch')['math score'].mean().plot(kind='bar')
plt.title("Lunch Type vs Math Score")
plt.ylabel("Average Score")
plt.show()
```

```python
import numpy as np

corr = df[['math score','reading score','writing score']].corr().values

plt.figure(figsize=(6,4))
plt.imshow(corr, cmap='coolwarm')
plt.colorbar()

labels = ['Math', 'Reading', 'Writing']
plt.xticks(np.arange(len(labels)), labels)
plt.yticks(np.arange(len(labels)), labels)

# Show correlation values
for i in range(len(labels)):
    for j in range(len(labels)):
        plt.text(j, i, f"{corr[i, j]:.2f}", ha='center', va='center', color='black')

plt.title("Correlation Between Scores (Matplotlib Only)")
plt.show()
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.hist(df['math score'], color='red')
plt.title("Math Score")

plt.subplot(1,3,2)
plt.hist(df['reading score'], color='green')
plt.title("Reading Score")

plt.subplot(1,3,3)
plt.hist(df['writing score'], color='blue')
plt.title("Writing Score")

plt.show()
```
